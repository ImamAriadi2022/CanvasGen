## 3. Diagnosa Sistem & Hardware Lokal
Memeriksa versi Python, driver CUDA, metadata perangkat GPU, status PyTorch CUDA, RAM Sistem, dan ruang Disk.

```python
import sys
import shutil

try:
    import psutil
    PSUTIL_OK = True
except ImportError:
    PSUTIL_OK = False

try:
    import torch
    TORCH_OK = True
except ImportError:
    TORCH_OK = False

print("=========================================================")
print("                LAPORAN DIAGNOSA SISTEM                  ")
print("=========================================================")
print(f"• Versi Python       : {sys.version.split()[0]}")

if TORCH_OK:
    print(f"• Versi PyTorch      : {torch.__version__}")
    print(f"• Status CUDA        : {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"• Versi CUDA         : {torch.version.cuda}")
        print(f"• Nama Perangkat GPU : {torch.cuda.get_device_name(0)}")
        vram_gb = torch.cuda.get_device_properties(0).total_memory / (1024**3)
        print(f"• Total VRAM         : {vram_gb:.2f} GB")
    else:
        print("ℹ️ Akselerasi CUDA GPU tidak terdeteksi (Mode CPU).")
else:
    print("ℹ️ PyTorch belum terpasang (Mode Engine Lightweight/Mock).")

# Memori RAM Sistem & Ruang Disk
if PSUTIL_OK:
    ram_gb = psutil.virtual_memory().total / (1024**3)
    print(f"• RAM Sistem         : {ram_gb:.2f} GB")

disk_total, disk_used, disk_free = shutil.disk_usage('/')
print(f"• Sisa Ruang Disk    : {disk_free / (1024**3):.2f} GB")
print("=========================================================")
```

## 4. Verifikasi Struktur Proyek & Unit Test
Jalankan pengujian otomatis untuk memverifikasi seluruh modul CanvasGen secara lokal.

```python
# Jalankan pengujian otomatis CanvasGen
import unittest
import sys

sys.path.insert(0, '.')
print("Menjalankan pengujian otomatis CanvasGen...")
suite = unittest.defaultTestLoader.discover('tests')
runner = unittest.TextTestRunner(verbosity=1)
res = runner.run(suite)

if res.wasSuccessful():
    print("✅ Seluruh pengujian unit CanvasGen LULUS 100%!")
else:
    print("❌ Pengujian mendeteksi kegagalan.")
```
